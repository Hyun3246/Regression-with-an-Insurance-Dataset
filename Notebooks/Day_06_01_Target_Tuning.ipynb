```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sample_data/01_train.csv')
```

```python
df['Premium Amount'].hist(bins=50, figsize=(20,15))
```

```python
# Premium Amount has a long tail -> log transformer
from sklearn.preprocessing import FunctionTransformer
import numpy as np

log_transformer = FunctionTransformer(np.log, inverse_func=np.exp)
df['Premium Amount'] = log_transformer.fit_transform(df[['Premium Amount']])
```

```python
import matplotlib.pyplot as plt

df['Annual Income'].hist(bins=50, figsize=(20,15))
plt.show()
```

```python
df.to_csv('01_train_log_target.csv', index=False)
```
